Imprima los números primos hasta un cierto n

```python
import numpy as np

def primos(x):
    numeros = list(range(2,x+1))
    d=0
    while numeros[d]**2 <= x:
        for n in numeros:
            if n != numeros[d]:
                if n % numeros[d] == 0:
                    numeros.remove(n)
        d += 1
    return numeros

primos(200)
```

Determine si una cadena de texto dada es un isograma, es decir, no se repite ninguna letra.
Ejemplos válidos de isogramas:
• lumberjacks
• background
• downstream
• six-year-old

```python
word = list(input('Ingrese texto\n'))
copy = []

for i in range(len(word)):
    if word[i] not in copy:
        copy.append(word[i])

if len(copy) != len(word):
    print('not an isogram')
else:
    print('is isogram')

```

Escriba una clase MobilePhone que represente un teléfono móvil.
Atributos:
• manufacturer (cadena de texto)
• screen_size (flotante)
• num_cores (entero)
• apps (lista de cadenas de texto)
• status (0: apagado, 1: encendido)
Métodos:
• __init__(self, manufacturer, screen_size, num_cores)
• power_on(self)
• power_off(self)
• install_app(self, app)
• uninstall_app(self, app)
6.2. Objetos y Clases 259
Aprende Python
Crear al menos una instancia (móvil) a partir de la clase creada y «jugar» con los métodos,
visualizando cómo cambian sus atributos.
¿Serías capaz de extender el método install_app() para instalar varias aplicaciones a la
vez?

```python
class MobilePhone:

    def __init__(self, manufacturer = 'Chayomi', screen_size = 6.3, num_cores = 8, apps = ['fb'], status = False):
        self.manufacturer = manufacturer
        self.screen_size = screen_size
        self.num_cores = num_cores
        self.apps = apps
        self.status = status
        
    def show_info(self):
        print(self.manufacturer, self.screen_size, self.num_cores, self.apps, self.status)
    
    def power_on(self):
        self.status = True
        print("Encendido")
        
    def power_off(self):
        self.status = False
        print("Encendido") 
        
    def install_app(self, *app):
        for i in range(len(app)):
            self.apps.append(app[i])
            
    def uninstall_app(self, app):
        self.apps.pop(app)
    
        
```

```python
cel1 = MobilePhone()
cel1.install_app('ig','tw','tg')
cel1.show_info()
```
