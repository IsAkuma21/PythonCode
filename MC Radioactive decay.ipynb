```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N0 = 5000
half_life = 110
l_decay = np.log(2) / half_life
dt = 1/2/half_life
t0 = 0

N = []
T = []

for i in range (500000):
    p = np.random.rand()
    if p < N0*l_decay*dt:
        N0 -= 1
        t0 += dt
    else:
        t0 += dt
    N.append(N0)
    T.append(t0)
```

```python
plt.plot(T,N)
plt.show()
```

```python
N0 = 0
half_life = 110
l_decay = np.log(2) / half_life
dt = 1/2/half_life
t0 = 0
R = 1

N = []
T = []


for i in range (500000):
    p = np.random.rand()
    if p < N0*l_decay*dt:
        N0 -= 1
        t0 += dt
    if p < R*dt:
        N0 += 1
        t0 += dt
    else:
        t0 += dt
    N.append(N0)
    T.append(t0)
```

```python
plt.plot(T,N)
plt.show()
```

```python
N0_Mo = 10000
N0_Tc = 0
h_l_Mo = 66
h_l_Tc = 6

l_decay_Mo = np.log(2) / h_l_Mo
l_decay_Tc = np.log(2) / h_l_Tc

dt_mo = 1/2/h_l_Mo
dt_tc = 1/2/h_l_Tc
t0 = 0

N_Mo = []
N_Tc = []
T = []


for i in range (100000):
    p1 = np.random.rand()
    if p1 < N0_Mo*l_decay_Mo*dt_mo:
        N0_Mo -= 1
        N0_Tc += 1
        
        p2 = np.random.rand()
        if p2 < N0_Tc*l_decay_Tc*dt_mo:
            N0_Tc -= 1
        t0 += dt
    else:
        
        p2 = np.random.rand()
        if p2 < N0_Tc*l_decay_Tc*dt_mo:
            N0_Tc -= 1
        t0 += dt
        
    N_Mo.append(N0_Mo)
    N_Tc.append(N0_Tc)
    T.append(t0)
    
A_Mo = [elements*l_decay_Mo for elements in N_Mo]
A_Tc = [elements*l_decay_Tc for elements in N_Tc]
```

```python
plt.plot(T,A_Mo)
plt.plot(T,A_Tc)
plt.show()
```
